```python
# Mengimplementasikan langkah-langkah regresi linier ke dalam program Python.

import numpy as np

# Data bulan dan penjualan
X = np.array([1, 2, 3, 4, 5])  # Bulan
Y = np.array([220, 250, 210, 300, 325])  # Penjualan

# Menghitung nilai slope (b) dan intercept (a) menggunakan metode least squares
n = len(X)
sum_X = np.sum(X)
sum_Y = np.sum(Y)
sum_XY = np.sum(X * Y)
sum_X2 = np.sum(X ** 2)

# Menghitung b (slope) dan a (intercept)
b = (n * sum_XY - sum_X * sum_Y) / (n * sum_X2 - sum_X**2)
a = (sum_Y - b * sum_X) / n

# Fungsi regresi linier
def predict(x):
    return a + b * x

# Menghitung peramalan berdasarkan persamaan regresi
forecast = predict(X)

# Menghitung MAPE (Mean Absolute Percentage Error)
mape = np.mean(np.abs((Y - forecast) / Y)) * 100

# Menampilkan hasil perhitungan
a, b, forecast, mape
```

```python
import matplotlib.pyplot as plt

# Membuat plot data aktual dan data peramalan
plt.figure(figsize=(10, 6))

# Plot penjualan aktual
plt.plot(X, Y, label='Penjualan Aktual', color='blue', marker='o', linestyle='--')

# Plot hasil peramalan
plt.plot(X, forecast, label='Peramalan', color='red', marker='x', linestyle='-')

# Memberi label pada sumbu dan judul grafik
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.title('Perbandingan Penjualan Aktual dan Hasil Peramalan')

# Menampilkan legenda
plt.legend()

# Menampilkan grid
plt.grid(True)

# Menampilkan grafik
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor

# Data yang diberikan sebelumnya (bulan 1 hingga 5)
bulan_actual = np.array([1, 2, 3, 4, 5])
penjualan_actual = np.array([220, 250, 210, 300, 325])

# Memisahkan data latih dan uji
X_actual = bulan_actual.reshape(-1, 1)  # Ubah data bulan menjadi 2D array
Y_actual = penjualan_actual

# Membuat model Random Forest Regressor
model_rf_actual = RandomForestRegressor(n_estimators=100, random_state=42)
model_rf_actual.fit(X_actual, Y_actual)

# Melakukan prediksi
Y_pred_actual = model_rf_actual.predict(X_actual)

# Visualisasi hasil prediksi vs penjualan aktual
plt.figure(figsize=(10, 6))

# Plot data aktual
plt.plot(bulan_actual, penjualan_actual, label='Penjualan Aktual', color='blue', marker='o', linestyle='--')

# Plot prediksi Random Forest
plt.plot(bulan_actual, Y_pred_actual, label='Peramalan (Random Forest)', color='green', marker='x', linestyle='-')

# Memberi label pada sumbu dan judul grafik
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.title('Perbandingan Penjualan Aktual dan Hasil Peramalan (Random Forest)')

# Menampilkan legenda
plt.legend()

# Menampilkan grid
plt.grid(True)

# Menampilkan grafik
plt.show()

# Menampilkan hasil prediksi
Y_pred_actual
```
