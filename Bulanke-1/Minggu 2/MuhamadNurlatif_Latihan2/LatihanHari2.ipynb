```python
#groupby = mengroup data per orang sekaligus salesnya dijumlahkan
top_c = df[['Customer Name', 'Sales']].groupby(["Customer Name"]).sum()

#sortvalues = mengurutkan berdasarkan kolom tertentu
#ascending=false = mengurut dari besar ke terkecil
top_c = top_c.sort_values("Sales", ascending=False)

#head(20) = mengambil 20 data pertama dari atas
top_c = top_c.head(20)

#round() = formating nilai di belakang koma
top_c = top_c[["Sales"]].round(2)

#
top_c.reset_index(inplace = True)
top_c
```

```python
plt.figure(figsize = (15,5))


#menggambar grafik dengan warna green
#menuliskan judul grafik dengan ukuran font 18
plt.title("Most Valuable Customers (2015-2019)", fontsize = 18)
plt.bar(top_c["Customer Name"],
       top_c["Sales"], color = '#99ff99',edgecolor = 'green',
       linewidth = 1)
plt.xlabel("Customers", fontsize= 15) #label sumbu x
plt.ylabel("Revenue",fontsize = 15) #label sumbu y
plt.xticks(fontsize= 12,rotation = 90) #label sumbu x dirotasi 90 derajat
plt.yticks(fontsize = 12)
for k,v in top_c ["Sales"].items():
    plt.text(k,v-8000,'$'+str(v),fontsize = 12, rotation=90,color='k',
            horizontalalignment='center');
```

```python
import seaborn as sns

newdf = df.copy()

newdf.set_index('Order Date', inplace=True)

monthly_s=newdf['Sales'].resample('ME').sum()

plt.figure(figsize=(12,6))
sns.lineplot(x=monthly_s.index, y=monthly_s.values)
plt.title('Monthly Sales Trend - Sample Superstore')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()

yearly_s = newdf['Sales'].resample('YE').sum()
plt.figure(figsize = (10,5))
sns.barplot(x=yearly_s.index.year, y=yearly_s.values, palette='viridis', legend = False)
plt.title('Yearly Sales Trend')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```
