```python
import tkinter as tk

def latihan_3():
    root = tk.Tk()
    root.title("Latihan 3: Tombol Klik")
    
    label = tk.Label(root, text="Klik tombol di bawah")
    label.pack()

    # Fungsi untuk mengubah teks label saat tombol diklik
    def ganti_teks():
        label.config(text="Tombol Sudah Diklik")

    # Membuat tombol di luar fungsi ganti_teks
    tombol = tk.Button(root, text="Klik Saya", command=ganti_teks)
    tombol.pack()

    root.mainloop()

latihan_3()
```

```python
import tkinter as tk

def latihan_5():
    root = tk.Tk()
    root.title("Latihan 5: Radio Button")
    
    # Variabel untuk menyimpan pilihan
    pilihan = tk.StringVar(value="")
    
    # Fungsi untuk menampilkan pilihan
    def tampilkan_pilihan():
        hasil.config(text=f"Kamu memilih: {pilihan.get()}")
    
    # Radio Button 1
    rb1 = tk.Radiobutton(root, text="Pilihan 1", variable=pilihan, value="Pilihan 1", command=tampilkan_pilihan)
    rb1.pack()
    
    # Radio Button 2
    rb2 = tk.Radiobutton(root, text="Pilihan 2", variable=pilihan, value="Pilihan 2", command=tampilkan_pilihan)
    rb2.pack()
    
    # Label untuk menampilkan hasil pilihan
    hasil = tk.Label(root, text="")
    hasil.pack()
    
    root.mainloop()

latihan_5()
```

```python
import tkinter as tk

def latihan_bilangan_prima():
    root = tk.Tk()
    root.title("Latihan: Cek Bilangan Prima")
    
    # Fungsi untuk mengecek apakah bilangan prima
    def cek_prima():
        try:
            angka = int(entry_angka.get())
            if angka > 1:
                for i in range(2, angka):
                    if angka % i == 0:
                        label_hasil.config(text=f"{angka} bukan bilangan prima.")
                        break
                else:
                    label_hasil.config(text=f"{angka} adalah bilangan prima.")
            else:
                label_hasil.config(text=f"{angka} bukan bilangan prima.")
        except ValueError:
            label_hasil.config(text="Input tidak valid, masukkan angka.")

    # Input untuk angka yang ingin diperiksa
    tk.Label(root, text="Masukkan Angka:").pack()
    entry_angka = tk.Entry(root)
    entry_angka.pack()

    # Tombol untuk cek prima
    tombol_cek = tk.Button(root, text="Cek Bilangan Prima", command=cek_prima)
    tombol_cek.pack()

    # Label untuk menampilkan hasil
    label_hasil = tk.Label(root, text="Hasil: ")
    label_hasil.pack()

    root.mainloop()

latihan_bilangan_prima()
```

```python
def latihan_9():
    root = tk.Tk()
    root.title("Latihan 9: Skala")
    def update_value(value):
        label_value.config(text=f"Nilai: {value}")

    scale = tk.Scale(root, from_=0, to = 100, orient=tk.HORIZONTAL, command=update_value)
    scale.pack()

    label_value = tk.Label (root, text = "Nilai: 0")
    label_value.pack()

    root.mainloop()

latihan_9()
```

```python
import tkinter as tk
from tkinter import ttk

def latihan_10():
    root = tk.Tk()
    root.title("Latihan 10: Tabel Data")

    tree = ttk.Treeview(root, columns=("Nama", "Umur", "Kota"), show='headings')

    tree.heading("Nama", text="Nama")
    tree.heading("Umur", text="Umur")
    tree.heading("Kota", text="Kota")

    data = [
        ("Budi", 25, "Jakarta"),
        ("Susi", 30, "Bandung"),
        ("Andi", 22, "Surabaya"),
    ]
    
    for row in data:
        tree.insert("", tk.END, values=row)

    tree.pack()

    root.mainloop()

latihan_10()
```

```python
import tkinter as tk

def latihan_11():
    root = tk.Tk()
    root.title("Latihan 11: To-Do List")

    # Daftar untuk menyimpan tugas
    tasks = []

    def tambah_tugas():
        tugas = entry_tugas.get()
        if tugas:
            tasks.append(tugas)
            update_listbox()
            entry_tugas.delete(0, tk.END)

    def hapus_tugas():
        selected_task = listbox_tugas.curselection()
        if selected_task:
            tasks.pop(selected_task[0])
            update_listbox()

    def update_listbox():
        listbox_tugas.delete(0, tk.END)
        for task in tasks:
            listbox_tugas.insert(tk.END, task)


    entry_tugas = tk.Entry(root)
    entry_tugas.pack()

    tombol_tambah = tk.Button(root, text="Tambah Tugas", command=tambah_tugas)
    tombol_tambah.pack()

    tombol_hapus = tk.Button(root, text="Hapus Tugas", command=hapus_tugas)
    tombol_hapus.pack()

    listbox_tugas = tk.Listbox(root)
    listbox_tugas.pack()

    root.mainloop()

latihan_11()
```

```python
import tkinter as tk
from tkinter import messagebox

def latihan_13():
    root = tk.Tk()
    root.title("Latihan 13: Buku Telepon")

    kontak = []

    def tambah_kontak():
        nama = entry_nama.get()
        nomor = entry_nomor.get()
        if nama and nomor:
            kontak.append((nama, nomor))
            update_listbox()
            entry_nama.delete(0, tk.END)
            entry_nomor.delete(0, tk.END)
        else:
            messagebox.showwarning("Input Error", "Nama dan Nomor harus diisi!")

    def hapus_kontak():
        selected_index = listbox_kontak.curselection()
        if selected_index:
            kontak.pop(selected_index[0])
            update_listbox()
        else:
            messagebox.showwarning("Pilih Kontak", "Pilih kontak yang akan dihapus!")

    def update_listbox():
        listbox_kontak.delete(0, tk.END)
        for nama, nomor in kontak:
            listbox_kontak.insert(tk.END, f"{nama} - {nomor}")

    tk.Label(root, text="Nama:").pack()
    entry_nama = tk.Entry(root)
    entry_nama.pack()

    tk.Label(root, text="Nomor Telepon:").pack()
    entry_nomor = tk.Entry(root)
    entry_nomor.pack()

    tombol_tambah = tk.Button(root, text="Tambah Kontak", command=tambah_kontak)
    tombol_tambah.pack()

    tombol_hapus = tk.Button(root, text="Hapus Kontak", command=hapus_kontak)
    tombol_hapus.pack()

    listbox_kontak = tk.Listbox(root)
    listbox_kontak.pack()

    root.mainloop()

latihan_13()
```
