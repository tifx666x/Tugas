```python
import mysql.connector

# Membuat koneksi ke database MySQL
def get_connection():
    return mysql.connector.connect(
        host="localhost",      
        user="root",           
        password="",   
        database="perpustakaandigital"
    )

# CREATE: Menambah data baru ke tabel pelanggan
def create_pelanggan(nama, alamat):
    conn = get_connection()
    cursor = conn.cursor()
    try:
        query = "INSERT INTO pelanggan (Nama_Pelanggan, Alamat) VALUES (%s, %s)"
        cursor.execute(query, (nama, alamat))
        conn.commit()
        print(f"Data pelanggan {nama} berhasil ditambahkan.")
    except mysql.connector.Error as err:
        print(f"Error: {err}")
    finally:
        cursor.close()
        conn.close()

# READ: Menampilkan semua data dari tabel pelanggan
def read_pelanggan():
    conn = get_connection()
    cursor = conn.cursor()
    try:
        query = "SELECT * FROM pelanggan"
        cursor.execute(query)
        results = cursor.fetchall()

        print("Data Pelanggan:")
        for row in results:
            print(f"ID: {row[0]}, Nama: {row[1]}, Alamat: {row[2]}")
    except mysql.connector.Error as err:
        print(f"Error: {err}")
    finally:
        cursor.close()
        conn.close()

# UPDATE: Mengubah data pelanggan berdasarkan id
def update_pelanggan(id_pelanggan, nama_baru, alamat_baru):
    conn = get_connection()
    cursor = conn.cursor()
    try:
        query = "UPDATE pelanggan SET Nama_Pelanggan = %s, Alamat = %s WHERE ID_pelanggan = %s"
        cursor.execute(query, (nama_baru, alamat_baru, id_pelanggan))
        conn.commit()
        print(f"Data pelanggan dengan ID {id_pelanggan} berhasil diubah.")
    except mysql.connector.Error as err:
        print(f"Error: {err}")
    finally:
        cursor.close()
        conn.close()

# DELETE: Menghapus data pelanggan berdasarkan id
def delete_pelanggan(id_pelanggan):
    conn = get_connection()
    cursor = conn.cursor()
    try:
        query = "DELETE FROM pelanggan WHERE ID_Pelanggan = %s"
        cursor.execute(query, (id_pelanggan,))
        conn.commit()
        print(f"Data pelanggan dengan ID {id_pelanggan} berhasil dihapus.")
    except mysql.connector.Error as err:
        print(f"Error: {err}")
    finally:
        cursor.close()
        conn.close()

# Menjalankan fungsi CRUD
if __name__ == "__main__":
    # Menambah data (CREATE)
    create_pelanggan("Jonathan", "Jl.  No. 10")

    # Menampilkan semua data (READ)
    read_pelanggan()

    # Mengubah data (UPDATE)


    # Menampilkan semua data setelah update
    read_pelanggan()

    # Menghapus data (DELETE)
    delete_pelanggan(2)

    # Menampilkan semua data setelah delete
    read_pelanggan()
```

```python
import mysql.connector

# Membuat koneksi ke database MySQL
def get_connection():
    return mysql.connector.connect(
        host="localhost",      
        user="root",           
        password="",   
        database="perpustakaandigital"
    )

# CREATE: Menambah data baru ke tabel buku
def create_buku(judul_buku, pengarang, tahun_terbit, kategori, jumlah):
    conn = get_connection()
    cursor = conn.cursor()
    try:
        query = "INSERT INTO buku (Judul_Buku, Pengarang, Tahun_Terbit, Kategori, Jumlah) VALUES (%s, %s, %s, %s, %s)"
        cursor.execute(query, (judul_buku, pengarang, tahun_terbit, kategori, jumlah))
        conn.commit()
        print(f"Data buku {judul_buku} berhasil ditambahkan.")
    except mysql.connector.Error as err:
        print(f"Error: {err}")
    finally:
        cursor.close()
        conn.close()

# READ: Menampilkan semua data dari tabel buku
def read_buku():
    conn = get_connection()
    cursor = conn.cursor()
    try:
        query = "SELECT * FROM buku"
        cursor.execute(query)
        results = cursor.fetchall()

        print("Data Buku:")
        for row in results:
            print(f"ID: {row[0]}, Judul: {row[1]}, Pengarang: {row[2]}, Tahun: {row[3]}, Kategori: {row[4]}, Jumlah: {row[5]}")
    except mysql.connector.Error as err:
        print(f"Error: {err}")
    finally:
        cursor.close()
        conn.close()

# UPDATE: Mengubah data buku berdasarkan id
def update_buku(id_buku, judul_baru, pengarang_baru, tahun_terbit_baru, kategori_baru, jumlah_baru):
    conn = get_connection()
    cursor = conn.cursor()
    try:
        query = "UPDATE buku SET Judul_Buku = %s, Pengarang = %s, Tahun_Terbit = %s, Kategori = %s, Jumlah = %s WHERE ID_Buku = %s"
        cursor.execute(query, (judul_baru, pengarang_baru, tahun_terbit_baru, kategori_baru, jumlah_baru, id_buku))
        conn.commit()
        print(f"Data buku dengan ID {id_buku} berhasil diubah.")
    except mysql.connector.Error as err:
        print(f"Error: {err}")
    finally:
        cursor.close()
        conn.close()

# DELETE: Menghapus data buku berdasarkan id
def delete_buku(id_buku):
    conn = get_connection()
    cursor = conn.cursor()
    try:
        query = "DELETE FROM buku WHERE ID_Buku = %s"
        cursor.execute(query, (id_buku,))
        conn.commit()
        print(f"Data buku dengan ID {id_buku} berhasil dihapus.")
    except mysql.connector.Error as err:
        print(f"Error: {err}")
    finally:
        cursor.close()
        conn.close()
```
